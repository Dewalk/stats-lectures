```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = '1972 190 1973 186 1974 167 1975 162 1976 167 1977 168 1978 165 1979 155 1980 154 1981 147 1982 150 1983 143 1984 149 1985 154 1986 145 1987 146 1988 145 1989 144 1990 145 1991 144 1992 144 1993 145 1994 142 1995 145 1996 147 1997 146 1998 143 1999 143 2000 146 2001 144 2002 141 2003 145 2004 144 2005 145 2006 143 2007 149 2008 145 2009 152'
```

```python
data = data.split()
year = []
times = []

for item in data:
    val = int(item)
    if val >= 1971:
        year.append(val)
    else:
        times.append(val)
print(year)
print(times)
```

```python
df = pd.DataFrame(zip(year, times), columns=['year', 'time'])
```
